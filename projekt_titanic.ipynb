Gibt es DatenlÃ¼cken in Dataset? Wie kann man sie behandeln?

```python
df_lucken = df.isnull().sum()
df_lucken
```

```python
df_lucken = df.isna().sum()

df['Age'].fillna(value=df['Age'].mean(), inplace=True)
df_lucken = df.isna().sum()
df_lucken
```

Gibt es kategorische Attribute in Dataset? Kann man diese in Zahlen umwandeln?

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
# df['Sex'] = df['Sex'].replace({'male': 0, 'female': 1})
# One-Hot Encoding
df = pd.get_dummies(df, columns=['Embarked', 'Pclass']) #drop_first=True)
```

```python
df.info()
```

Sind alle Datentypen in Dataset _richtig_ und fÃ¼r analytische Aufgaben geeignet? (FÃ¼r nummerische Analysen mÃ¼ssen die Daten als Zahlen vorliegen, welche Spalte kÃ¶nnte man entsprechend umwandeln ? ) <br>
Zeicchenketten kÃ¶nnen unter anderem der `replace()`-Methode umgewandelt werden, in dem man jedes Vorkommen durch eine entsprechende Zahl ersetzt. 

```python
df['Cabin'] = df['Cabin'].fillna('U').apply(lambda x: x[0])  # 'U' fÃ¼r unbekannt
df['Cabin'] = df['Cabin'].replace({c: i for i, c in enumerate(df['Cabin'].unique())})
df.head()
```

Wie viele von Passagieren waren weiblich und wie viele mÃ¤nnlich?

```python
geschlecht = df.groupby('Sex')['Sex'].size()
geschlecht.to_frame('Number')
```

Wie ist der Prozentanteil der geretteten weiblichen Personen im Vergleich zu geretteten mÃ¤nnlichen Personen?

```python
survived_female = df[(df['Survived'] == 1) & (df['Sex'] == 1)].shape[0]
survived_male = df[(df['Survived'] == 1) & (df['Sex'] == 0)].shape[0]

total_survived = survived_female + survived_male
percent_survived_female = (survived_female / total_survived) * 100
percent_survived_male = (survived_male / total_survived) * 100

percent_survived_female, percent_survived_male
```

Wie ist der Durchschnittalter von weiblichen Personen, bzw. mÃ¤nnlichen Passagieren?

```python
average_age_female = df[df['Sex'] == 1]['Age'].mean()
average_age_male = df[df['Sex'] == 0]['Age'].mean()
average_age_female.round(2), average_age_male.round(2)
```
